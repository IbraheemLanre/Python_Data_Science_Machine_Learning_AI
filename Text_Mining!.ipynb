```python
import nltk
nltk.download()

# Simple Sentiment Analysis. Computing the proportion of positive and negative words in a text. Two restaurants reviews
def get_words(url):
    import requests
    words = requests.get(url).content.decode('latin-1')
    word_list = words.split('\n')
    index = 0
    while index < len(word_list):
        word = word_list[index]
        if ';' in word or not word:
            word_list.pop(index)
        else:
            index += 1
    return word_list

p_url = 'http://ptrckprry.com/course/ssd/data/positive-words.txt'
n_url = 'http://ptrckprry.com/course/ssd/data/negative-words.txt'
positive_words = get_words(p_url)
negative_words = get_words(n_url)

len(positive_words)
len(negative_words)
```

```python
# Computing sentiment by looking at the proportion of positive and negative words in the text

from nltk import word_tokenize
with open('data/community.txt', 'r') as f:
    community = f.read()
with open ('data/le_monde.txt', 'r') as f:
    le_monde = f.read()

from nltk import word_tokenize
cpos = cneg = lpos = lneg = 0
for word in word_tokenize(community):
    if word in positive_words:
        cpos += 1
    if word in negative_words:
        cneg += 1
for word in word_tokenize(le_monde):
    if word in positive_words:
        lpos += 1
    if word in negative_words:
        lneg += 1
print('community {0:1.2f}%\t {1:1.2f}%\ {2:1.2f}%'.format(cpos/len(word_tokenize(community))*100, 
                                                         cneg/len(word_tokenize(community))*100,
                                                         (cpos-cneg)/len(word_tokenize(community))*100))
print('le_monde {0:1.2f}%\t {1:1.2f}%\ {2:1.2f}%'.format(lpos/len(word_tokenize(le_monde))*100,
                                                         lneg/len(word_tokenize(le_monde))*100,
                                                      (lpos-lneg)/len(word_tokenize(le_monde))*100))
```
